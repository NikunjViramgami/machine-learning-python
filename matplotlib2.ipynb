```python
categories = np.array(['A','B','C','D'])
x = np.array([10,20,30,40])
y = np.array([5,15,25,35])
z = np.array([2,22,23,25])
bar_width = 0.50
plt.bar(categories,x,bar_width,label='Dataset1')
plt.bar(categories,y,bar_width-0.10,label='Dataset2')
plt.bar(categories,z,bar_width-0.20,label='Dataset3')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Demo graph')
plt.legend()
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 120)#starring at a angles specify
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0.2]
#highlight the individual portion

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
#shadow describe shadow
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]
#describe colors 
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
#describe percentage of every parts
plt.pie(y, labels = mylabels,autopct='%1.1f%%')
plt.legend() #describe box of items
plt.show() 
```
